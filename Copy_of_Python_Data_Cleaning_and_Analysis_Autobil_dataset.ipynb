```
#The simplest pivot table must have a dataframe and an index . 
#In this case, let’s use the new column Date but specifically working on the hour...dt.hour as our index.
#You can have multiple indexes as well. In fact, most of the pivot_table args can take multiple values via a list.
#we will use one in ours
#to select paris in column city we use 
#For this purpose, the utilib and utilib 1.4  columns aren’t really useful.
# Let’s remove it by explicitly defining the columns we care about using the values field which is bluecars counter.
#The bluecar column automatically averages the data but we can do a count or a sum. Adding them is simple using aggfunc sum
#the busiest hour is the 21st hour
df2[df2['City']=='Paris'].pivot_table(index=df2['Date'].dt.hour ,values = "Bluec" ,aggfunc ='sum',margins=True)


```
